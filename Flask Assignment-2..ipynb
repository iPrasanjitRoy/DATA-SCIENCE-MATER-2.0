```python
"""
Q1. Explain GET and POST methods.

The GET and POST methods are HTTP request methods used in web development to send data from a client (such as a web browser) to a server.

GET method:

The GET method is used to request data from a server. It retrieves data specified in the URL's query parameters.
GET requests are limited in the amount of data they can send, as the data is appended to the URL.
GET requests should not be used for sensitive or confidential data, as the data is visible in the URL and can be logged or stored in browser history.



POST method:

The POST method is used to submit data to a server to create or update resources.
It sends the data in the body of the HTTP request, rather than in the URL like in a GET request.

POST requests can send larger amounts of data compared to GET requests, as the data is not limited by the URL length constraint.
POST requests are more secure for transmitting sensitive or confidential data, as the data is not exposed in the URL.

"""
```
